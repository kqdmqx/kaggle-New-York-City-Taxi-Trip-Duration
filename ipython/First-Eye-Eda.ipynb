```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
train = pd.read_csv('../input/train.csv', parse_dates=['pickup_datetime', 'dropoff_datetime'])
test = pd.read_csv('../input/test.csv', parse_dates=['pickup_datetime'])
train.head()
```

# year month distribution

```python
train['pickupYearMonth'] = train.pickup_datetime.dt.year * 100 + train.pickup_datetime.dt.month
test['pickupYearMonth'] = test.pickup_datetime.dt.year * 100 + test.pickup_datetime.dt.month

plt.figure(figsize=(12, 6))
train.pickupYearMonth.value_counts().sort_index().plot(kind='bar')
plt.title('train year month')
plt.show()

plt.figure(figsize=(12, 6))
test.pickupYearMonth.value_counts().sort_index().plot(kind='bar')
plt.title('test year month')
plt.show()
```

# duration in each month

```python
plt.figure(figsize=(12,6))
sns.boxplot(x="pickupYearMonth", y="trip_duration", data=train)
plt.ylabel('Median Duration', fontsize=12)
plt.xlabel('Pickup Year Month', fontsize=12)
plt.xticks(rotation='vertical')
plt.show()

plt.figure(figsize=(12,6))
plt.title('Over 1000000 Removed')
sns.boxplot(x="pickupYearMonth", y="trip_duration", data=train.loc[train.trip_duration < 1000000])
plt.ylabel('Median Duration', fontsize=12)
plt.xlabel('Pickup Year Month', fontsize=12)
plt.xticks(rotation='vertical')
plt.show()

plt.figure(figsize=(12,6))
plt.title('Over 10000 Removed')
sns.boxplot(x="pickupYearMonth", y="trip_duration", data=train.loc[train.trip_duration < 10000])
plt.ylabel('Median Duration', fontsize=12)
plt.xlabel('Pickup Year Month', fontsize=12)
plt.xticks(rotation='vertical')
plt.show()

plt.figure(figsize=(12,6))
plt.title('Over 1000 Removed')
sns.violinplot(x="pickupYearMonth", y="trip_duration", data=train.loc[train.trip_duration < 1000])
plt.ylabel('Median Duration', fontsize=12)
plt.xlabel('Pickup Year Month', fontsize=12)
plt.xticks(rotation='vertical')
plt.show()
```
